## Bivariate Analysis

```python
#sales by Category

plt.figure(figsize=(6,4))
sns.barplot(x='Category' , y='Sales' , data=df)
plt.title('Total Sales by Category')
plt.show()
```

Observation:
Technology generated the highest sales among all categories, followed by Furniture. Office Supplies recorded comparatively lower sales despite having high order frequency.

```python
#profit by region

plt.figure(figsize=(14,6))
sns.barplot(x='Region' , y='Profit' , data=df)
plt.title('Total Profit by Region')
plt.xticks(rotation=30)
plt.show()
```

Observation:
North Asia and Central Asia contributed the highest profit, while Southeast Asia and EMEA showed comparatively lower profitability.

```python
#Sales by Segments

plt.figure(figsize=(8,4))
sns.barplot(x='Segment' , y='Sales' , data=df)
plt.title(' Sales by Segments')
plt.show()
```

```python
#Discount vs profit

plt.figure(figsize=(7,4))
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Discount vs Profit')
plt.show()
```

Observation:
A negative relationship was observed between discount and profit, indicating that higher discounts often resulted in lower profit margins.

```python
#Profit by Category

plt.figure(figsize=(6,4))
sns.barplot(x='Category' , y='Profit' , data=df)
plt.title(' Profit by Category')
plt.show()
```

```python
#Sales by Region

plt.figure(figsize=(14,6))
sns.barplot(x='Region' , y='Profit' , data=df)
plt.title('Total Sales by Region')
plt.xticks(rotation=30)
plt.show()
```

```python
#Shipping Cost vs Profit

plt.figure(figsize=(12,6))
sns.scatterplot(x='Shipping Cost' , y='Profit' , data=df)
plt.title('Shipping Cost vs Profit')
plt.show()
```

## Correlation Analysis

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df[['Sales','Quantity','Discount','Profit','Shipping Cost']].corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')
plt.show()
```

Observation: 
Sales and Shipping Cost showed a strong positive correlation, while Discount and Profit showed a negative correlation. 
This suggests that larger discounts negatively impacted profitability.

## Multivariate Analysis

```python
sns.barplot(x='Category', y='Sales', hue='Segment', data=df)
```

```python
plt.figure(figsize=(13,6))
sns.barplot(x='Category', y='Sales', hue='Region', data=df)
plt.title(' Sales by Category and Region ')
plt.show()
```

Observation:

## Insights

1. Standard Class was the most preferred shipping method, while Same Day delivery was used the least, indicating that customers generally preferred cost-effective shipping options.

2. The Consumer segment contributed the highest number of customers, whereas the Home Office segment had comparatively lower participation.

3. The Central region recorded the highest order volume, while Canada contributed the lowest number of orders.

4. Office Supplies had the highest order frequency; however, Technology generated the highest sales and profit, showing that frequent purchases do not always contribute the most revenue.

5. Among product sub-categories, Binders and Storage were highly demanded, whereas Tables and Machines showed lower order frequency.

6. Medium-priority orders dominated overall transactions, while low-priority orders were the least common.

7. Most sales values were concentrated within the lower to medium range, while only a limited number of orders generated exceptionally high sales.

8. Profit values were primarily concentrated around smaller margins, although a few transactions resulted in significant profits as well as substantial losses.

9. Discounts were generally limited to lower ranges, but several transactions contained unusually high discount values, indicating the presence of outliers.

10. Customer purchases were mostly made in smaller quantities, with the majority of orders containing only a few units.

11. A negative relationship was observed between discount and profit, suggesting that higher discounts often reduced overall profitability.

12. Shipping cost did not show a consistent positive impact on profit, as some high shipping cost orders still resulted in financial losses.

13. Sales and Shipping Cost showed a strong positive correlation, indicating that larger sales transactions were usually associated with higher shipping expenses.

14. A moderate positive relationship was identified between Sales and Profit, showing that increased sales generally contributed to improved profitability.

15. Technology products consistently performed strongly across multiple customer segments and regions, while Office Supplies generated comparatively lower sales performance across most regions.

16. North Asia, Central Asia, and Oceania demonstrated stronger sales performance, particularly within the Technology and Furniture categories.


## Conclusion

The Exploratory Data Analysis (EDA) of the Global Superstore dataset helped identify important business trends, customer purchasing behavior, 
and sales performance patterns. 
The analysis revealed that Technology products generated the highest sales and profit, 
while Office Supplies had the highest order frequency but lower revenue contribution.

The study also showed that higher discounts often reduced profitability, and increased shipping cost did not always guarantee higher profit. 
Regional analysis highlighted that North Asia, Central Asia, and Oceania contributed strongly to overall sales performance. Overall, 
the analysis provided meaningful insights that can support better business decisions related to sales strategy, pricing, and operational efficiency.
